```python
kolumny = ["ID", "Age", "Gender", "Occupation", "ZIP Code"] #lista nazw kolumn jakie chcemy wprowadziƒá
movie = pd.read_table("http://bit.ly/movieusers", sep="|", header=None, names=kolumny)
movie.head() # wprowadzamy argument names= i przypisujemy do niego naszƒÖ liste z nazwami kolumn
```

```python
# pandas automatycznie przypisuje do tabeli sw√≥j index numeryczny, w tym przypadku za index mo≈ºe pos≈Çu≈ºyc nam
# kolumna ID (identyczne wartosci)
```

```python
movie.set_index("ID", inplace=True) # przypisanie kolumny ID jako index ( nalezy okreslic argument inplace= jako True je≈ºeli chcemy
                      # aby zmiany zosta≈Çy wprowadzone na stale, domyslnie jest inplace=False)
movie.head()    
```
