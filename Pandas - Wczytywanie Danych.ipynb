```python
movie = pd.read_table("http://bit.ly/movieusers", sep="|")
movie.head() # separator jest już ok, ale pierwszy wiersz jest uzyty jako naglowek i tzreba to zmienić (header=)
```

```python
movie = pd.read_table("http://bit.ly/movieusers", sep="|", header=None)
movie.head() # pozostaje tylko dodać nazwy kolumn(nagłowek)
```

```python
kolumny = ["ID", "Age", "Gender", "Occupation", "ZIP Code"] #lista nazw kolumn jakie chcemy wprowadzić
movie = pd.read_table("http://bit.ly/movieusers", sep="|", header=None, names=kolumny)
movie.head() # wprowadzamy argument names= i przypisujemy do niego naszą liste z nazwami kolumn
```

```python
# pandas automatycznie przypisuje do tabeli swój index numeryczny, w tym przypadku za index może posłużyc nam
# kolumna ID (identyczne wartosci)
```

```python
movie.set_index("ID", inplace=True) # przypisanie kolumny ID jako index ( nalezy okreslic argument inplace= jako True jeżeli chcemy
                      # aby zmiany zostały wprowadzone na stale, domyslnie jest inplace=False)
movie.head()    
```
